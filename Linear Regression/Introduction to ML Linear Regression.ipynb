```python
import pandas as pd

df_train = pd.read_csv('/Users/rohith/Documents/Datasets/Linear_Regression/train.csv')
df_test = pd.read_csv('/Users/rohith/Documents/Datasets/Linear_Regression/test.csv')
```

```python
import numpy as np

x_train = df_train['x']
y_train = df_train['y']
x_test = df_test['x']
y_test = df_test['y']

x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)
y_test = np.array(y_test)
```

```python
print(x_train.shape)
```

```python
x_train = x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score

clf = LinearRegression(normalize=True)
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
print(y_test[0])
print(y_pred[1])
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(10,10))
plt.scatter(range(len(y_test)),y_test,color='red',label='GT')
plt.plot(range(len(y_test)),y_pred,color='black',label = 'pred')
plt.legend()
plt.show()
```

```python
x_train = x_train.reshape(700,1)
```

```python
## Linear Regression 
import numpy as np

theta_0 = np.zeros((700,1))
theta_1 = np.zeros((700,1))


loops = 0
while(loops < 1000):
    y = theta_0 + theta_1 * x_train
    error = y - y_train
    mean_sq_er = np.sum(error**2)
    mean_sq_er = mean_sq_er/700
    if(abs(mean_sq_er) <= 0.01):
        break
    theta_0 = theta_0 - 0.00001 * 2 * np.sum(error)/700 
    theta_1 = theta_1 - 0.00001 * 2 * np.sum(error * x_train)/700
    loops += 1
    if(loops%10 == 0):
        print(mean_sq_er)
```

```python
x_test = x_test.reshape(300,1)
y_test = y_test.reshape(300,1)
```

```python
a_0 = theta_0[0]
a_1 = theta_1[0]
```

```python
print(a_1)
```

```python
y_prediction = a_0 + a_1 * x_test
```

```python
print(r2_score(y_test,y_prediction))
```
